```python
# Importing the neccessary libraries

import pandas as pd
```

```python
# Extraction layer

ziko_df = pd.read_csv(r'ziko_logistics_data.csv')
```

### Data cleaning and transformation

```python
for cols in ziko_df.columns:
    null_count = ziko_df[cols].isnull().sum()
    print(f'{cols} null values count is {null_count}')
```

```python
ziko_df.fillna({
    'Unit_Price': ziko_df['Unit_Price'].mean(),
    'Total_Cost':ziko_df['Total_Cost'].mean(),
    'Discount_Rate':0.0,
    'Return_Reason':'Unknown'
    },inplace=True
    )
ziko_df.info()
```

#### Convert the Data column from the string or Object data type to Datetime format

```python
# ziko_df['Date'] = ziko_df['Date'].astype('datetime64[ns]')
ziko_df['Date'] = pd.to_datetime(ziko_df['Date'])
ziko_df.info()
```

##### Create the customer table

```python
customer = ziko_df[['Customer_ID','Customer_Name','Customer_Phone','Customer_Email','Customer_Address']].copy()
customer.drop_duplicates(inplace=True)
```

```python
customer.dtypes
```

##### Create the product table

```python
product = ziko_df[['Product_ID','Product_List_Title','Quantity','Unit_Price']].copy()
product.drop_duplicates(inplace=True)
```

#### Merge_all the tables
- Merge is done inorder to fillout the fact_table 

```python
#transaction_merged = ziko_df.merge(customer,on=['Customer_ID','Customer_Name','Customer_Phone','Customer_Email','Customer_Address'],how='left')\
#      .merge(product,on=['Product_ID','Product_List_Title','Quantity','Unit_Price'],how='left')

transaction_merged = ziko_df.merge(customer,how='left').merge(product,how='left')
```

```python
# slice out the columns for the fact_table
transaction_fact = transaction_merged[['Transaction_ID', 'Date', 'Customer_ID', 'Product_ID','Total_Cost', 'Discount_Rate', 'Sales_Channel',
       'Order_Priority', 'Warehouse_Code', 'Ship_Mode', 'Delivery_Status',
       'Customer_Satisfaction', 'Item_Returned', 'Return_Reason',
       'Payment_Type', 'Taxable', 'Region', 'Country',]]
```

```python
transaction_fact
```

### Saving to tables csv

```python
customer.to_csv(r'customer.csv', index=False)
product.to_csv(r'product.csv',index=False)
transaction_fact.to_csv(r'transaction_fact.csv',index=False)
```
